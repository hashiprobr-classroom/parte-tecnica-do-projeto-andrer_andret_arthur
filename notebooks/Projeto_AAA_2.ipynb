```python
import csv
import netpixi
from netpixi.data.gt import *
import pandas as pd
from graph_tool import Graph, util, stats
g = Graph(directed=False)
import matplotlib.pyplot as plt
```

```python
games = []
genres = []
owners = []

with open('../notebooks/steam.csv') as file:
    reader = csv.reader(file, delimiter=',', quotechar='"')
    i = 0;
    for row in reader:
        game = row[1]
        genre = row[9]
        owner = row[16]
        games.append(game)
        genres.append(genre)
        owners.append(owner)
        i += 1
        
print(owners)
```

```python
splits = []
uniquecat = []
for o in range(len(genres)):
    unique = genres[o].split(";")
    splits.append(unique)

for u in range(len(splits)):
    for v in range(len(splits[u])):
        if splits[u][v] not in uniquecat:
            uniquecat.append(splits[u][v])
            
print(uniquecat)
```

```python
indexesToRemove = []
for gt in range(0,len(owners)):
    if owners[gt] == '0-20000' or owners[gt] == '20000-50000' or owners[gt] == '50000-100000' or owners[gt] =='100000-200000': #or owners[gt] == '200000-500000':
        indexesToRemove.append(gt)
print(len(indexesToRemove))
```

```python
games = [i for j, i in enumerate(games) if j not in indexesToRemove]
print(len(games))
```

```python
for all in range(len(games)):
    g.add_vertex()

print('número de vértices de g:', g.num_vertices())
```

```python
g.vp['name'] = g.new_vp('object')
for e in range(len(games)):
    vertice = g.vertex(e)
    g.vp['name'][vertice] = games[e]
```
